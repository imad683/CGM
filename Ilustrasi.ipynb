```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from matplotlib.colors import Normalize
from matplotlib import colormaps  # modern way
```

```python
# Define canvas size (height x width x RGB)
canvas_size = (30, 30, 3)  # 20x20 pixels, RGB image
canvas = np.ones(canvas_size)  # Start with a white canvas (1, 1, 1)

# Helper function to paint a pixel
def paint_pixel(x, y, color):
    canvas[y, x] = color  # y is row, x is column
```

```python
canvas = np.ones((30, 30, 3))
plt.figure(figsize=(8,8))
plt.rcParams['font.family'] = 'serif'

norm = Normalize(vmin=-300, vmax=300)
cmap = colormaps['RdYlBu_r']

# Galaxy
ellipticity = 0.81 
a = 3.3
b = a * (1 - ellipticity)

ellipse = Ellipse(xy=(16.785, 12.76), width=2*b, height=2*a, angle=-17.1,
                  edgecolor='black', facecolor=cmap(norm(0)))
plt.gca().add_patch(ellipse)

# Quasars
circle1 = Ellipse(xy=(19.5, 11.5), width=1, height=1, angle=0,
                  edgecolor='black', facecolor=cmap(norm(-191)))
circle2 = Ellipse(xy=(10.5, 17.5), width=1, height=1, angle=0,
                  edgecolor='black', facecolor='darkgrey')
plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)

plt.plot([3, 8],[3, 3], lw=1, color='black')
plt.text(5.5, 4, '7.30 kpc', fontsize=14, color='black', ha='center')
plt.text(9, 15.5, 'A', fontsize=16, color='black', horizontalalignment='center')
plt.text(20, 9, 'B', fontsize=16, color='black', horizontalalignment='center')
plt.text(17, 17, 'G', fontsize=16, color='black', ha='center')
plt.text(9, 14, '(Non-deteksi)', fontsize=12, color='black', ha='center')
plt.text(20, 7.7, '$v_r$ = −191.6 ± 0.4 km/s', fontsize=11, color='black', ha='center')

ax = plt.gca()
sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])
plt.colorbar(sm, ax=ax, label=r'$v_r$ [km/s]', shrink=0.83)

plt.imshow(canvas, interpolation='none')
plt.xlabel('Pixel', fontsize=16)
plt.ylabel('Pixel', fontsize=16)
plt.title('Kinematika Mg II Komponen 1', fontsize=16)
plt.gca().add_patch(ellipse)
plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)
plt.gca().invert_yaxis()
plt.axis('on')
plt.show()
```
